```python
n=30
a=1
b=1
Fibonacci=[]
#print(a)
#print(b)
Fibonacci.append(a)
Fibonacci.append(b)
for i in range(n):
    c=a+b
    #print(c)
    Fibonacci.append(c)
    a=b
    b=c
print(Fibonacci)
```
