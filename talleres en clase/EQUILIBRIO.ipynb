```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
deltaS_N2O4=304.376
deltaH_N2O4=9.079*1000    #se multiplica por 1000 porque esta en kilos
deltaS_NO2=240.034
deltaH_NO2=33.095*1000

T=298.15   #T en kelvin
pt=1.0    #presion total en bar
R=8.314  #j/k-mol
```

```python
uN2O4_est=deltaH_N2O4-T*deltaS_N2O4
uNO2_est=deltaH_NO2-T*deltaS_NO2
print(uN2O4_est,uNO2_est)
```

```python
x=np.arange(0.001,1,0.001)
n_N2O4=1-x       #x es shi (la E rara esa)
n_NO2=2*x
nt=1+x

X_N2O4=n_N2O4/nt
X_NO2=n_NO2/nt
```

```python
uN2O4=uN2O4_est+R*T*np.log(X_N2O4)
uNO2=uNO2_est+R*T*np.log(X_NO2)
uT=n_N2O4*uN2O4+n_NO2*uNO2


minimo=np.min(uT)
n=np.where(minimo==uT)
xeq=x[n]
print(xeq)


plt.plot(x,uT)
plt.axvline(x=xeq,color="red",linestyle="-.")
plt.text(xeq+0.02,-77000),r"$\
plt.xlabel(r"$\xi$")
plt.ylabel(r"$\Delta G$ (J/mol)")
plt.show()
```

```python
pNO2=2*xeq/(1+xeq)*pt
pN2O4=(1-xeq)/(1+xeq)*pt
Kp=pNO2**2/pN2O4
print(Kp)
```

```python
yy=(uN2O4-2*uNO2)**2
minimo=np.min(yy)
n=np.where(yy==minimo)
xeq=x[n]
print(xeq)

plt.plot(x,uN2O4)
plt.plot(x,2*uNO2)
plt.xlabel(r"$\xi$")
plt.ylabel(r"$\Delta G$ (J/mol)")
plt.axvline(x=xeq,color="red",linestyle="-.")
plt.show()
```
