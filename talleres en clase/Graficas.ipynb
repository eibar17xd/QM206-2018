```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([0,2,4,6],[3,1,7,4],color="fuchsia",linewidth=3,alpha=0.5,marker="o",markersize=15,markeredgewidth=2,markeredgecolor="black",markerfacecolor="deepskyblue")
plt.xlabel("$x(cm^2)$")
plt.ylabel("$intensidad(\mu)$")
plt.title("Experimento No. 1")
plt.grid(True)
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(4*x)
y2=np.cos(2*x)
plt.plot(x,y,color="limegreen")
plt.plot(x,y2,color="blue",dashes=[1,1,2,2])
plt.title("Función sen de 4x 0-$2\pi$")
plt.xlabel("t(s)")
plt.ylabel("Amplitud de la onda")
plt.grid(True)
plt.show()
```

```python
x=np.random.rand(100)*10
y=np.random.rand(100)*10
r=np.random.rand(100)*250+50
plt.scatter(x,y,s=r,color="teal",)
plt.show()
```

```python
x=np.random.randn(10000)
plt.hist(x,bins=50,facecolor="hotpink",edgecolor="white",density=True)
plt.title("Distribución de la pablación")
plt.ylabel("Densidad habitantes /$m^2$")
plt.text(-3.5, .35, r'$\mu=100, \ \sigma=15$')
plt.show()
```
