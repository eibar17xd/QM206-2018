```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## H2O GAS
A=30.092
B=6.832
C=6.793
D=-2.534448
E=0.082139

deltaH_298=-241.86
deltaS_298=188.835

dT=0.1
T=np.arange(300,400,dT)
t=T/1000
Cp_H2O_gas=A+B*t+C*t**2+D*t**3+E/t**2
deltaH_gas=np.cumsum(Cp_H2O_gas*dT)+deltaH_298*1000
deltaS_gas=np.cumsum(Cp_H2O_gas/T*dT)+deltaS_298
deltaG_gas=deltaH_gas-T*deltaS_gas
```

```python
## H2O LIQ
A=-203.606
B=1523.290
C=-3196.413
D=2474.455
E=3.855

deltaH_298=-285.830  #kJ/mol
deltaS_298=69.95       #kJ/mol

dT=0.1
T=np.arange(300,400,dT)
t=T/1000
Cp_H2O_liq=A+B*t+C*t**2+D*t**3+E/t**2
deltaH_liq=np.cumsum(Cp_H2O_liq*dT)+deltaH_298*1000
deltaS_liq=np.cumsum(Cp_H2O_liq/T*dT)+deltaS_298
deltaG_liq=deltaH_liq-T*deltaS_liq
```

```python
plt.plot(T,deltaG_gas)
plt.plot(T,deltaG_liq)
plt.show()
```

```python
dp=(deltaG_gas-deltaG_liq)**2
np.min(dp)   #buscar el minimo
minimo=np.min(dp)
np.where(dp==minimo)
```

```python
def leerArchivo(url):
    file=requests.get(url)
    datos=file.text
    datos=datos.split("\n")[3:-1]
    mat=[]
    for x in datos:
        s=x.split("\t")
        mat.append([float(a) for a in s])   #esto hace que en ves de que sea una lista de cadena lo cobierte en una lista de numero punto flotante
    mat=np.array(mat)
    return mat
```

```python
T=mat[:,0]
Cp=mat[:,1]
plt.plot(T,Cp)
plt.show()
```

```python
T=T[3:16]
y=Cp[3:16]
t=T/1000
n=len(T)
X=np.array([np.ones(n),t,t**2,t**3,1/t**2]).T
beta=np.linalg.inv(X.T.dot(X)).dot(X.T.dot(y))
beta
```

```python
Cp_calc=beta.dot(X.T)
```

```python
plt.plot(T,Cp_calc,color="orange")
plt.scatter(T,y)
plt.xlabel("T(K)")
plt.ylabel("Cp(J/K-mol)")
plt.show()
```
