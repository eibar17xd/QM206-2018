```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
wine=datasets.load_wine()
```

# wine datasets
Estos datos son el resultado de un análisis químico de vinos cultivados en la misma región en Italia pero derivados de tres cultivares diferentes. El análisis determinó las cantidades de 13 componentes encontrados en cada uno de los tres tipos de vinos.

Los atributos son
1) Alcohol 
2) Malic acid 
3) Ash 
4) Alcalinity of ash 
5) Magnesium 
6) Total phenols 
7) Flavanoids 
8) Nonflavanoid phenols 
9) Proanthocyanins 
10) Color intensity 
11) Hue 
12) OD280/OD315 of diluted wines 
13) Proline

```python
wine_dataframe=pd.DataFrame(data=wine.data,columns=wine.feature_names)
wine_dataframe["target"]=wine.target
x=wine_dataframe["ash"]
print(x.mean())
y=wine_dataframe["alcalinity_of_ash"]
col=(wine_dataframe["target"]+1)*2
print(y.mean())
ash=wine_dataframe["ash"]*60
alcalinity_of_ash=wine_dataframe["alcalinity_of_ash"]*60
```

# grafica 1
en esta grafica se observa la cantidad de cenizas Vs la alcalinidad de ceniza

```python
plt.scatter(x,y,c=col,s=ash,alpha=0.7)
plt.xlabel("cantidad de ceniza ")
plt.ylabel("alcalinidad de ceniza")
plt.title("cantidad de ceniza Vs alcalinidad de ceniza")
plt.grid(True)
plt.show()
```

# grafica 2

# grafica 3
en esta grafica se observa la cantidad de flavonoides Vs la cantidad de no flavonoides de fenoles

```python
wine_dataframe=pd.DataFrame(data=wine.data,columns=wine.feature_names)
wine_dataframe["target"]=wine.target
x=wine_dataframe["flavanoids"]
print(x.mean())
y=wine_dataframe["nonflavanoid_phenols"]
col=(wine_dataframe["target"]+1)*2
print(y.mean())
flavanoids=wine_dataframe["flavanoids"]*40
nonflavanoid_phenols=wine_dataframe["nonflavanoid_phenols"]*40
```
