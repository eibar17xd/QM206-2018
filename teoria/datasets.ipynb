```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris=datasets.load_iris()
iris.target_names
```

```python
iris_dataframe=pd.DataFrame(data=iris.data,columns=["SL","SW","PL","PW"])
iris_dataframe["target"]=iris.target
x=iris_dataframe["SL"]
print(x.mean())
y=iris_dataframe["SW"]
col=(iris_dataframe["target"]+1)*2
print(y.mean())
PL=iris_dataframe["PL"]*60
PW=iris_dataframe["PW"]*60
```

```python
plt.scatter(x,y,c=col,s=PL,alpha=0.7)
plt.xlabel("largo del spalo (cm)")
plt.ylabel("ancho del spalo (cm)")
plt.title("ancho de spalo Vs largo de spalo")
plt.grid(True)
plt.show()
```

```python
plt.scatter(x,y,c=col,s=PW,alpha=0.5)
plt.xlabel("largo del spalo (cm)")
plt.ylabel("ancho del spalo (cm)")
plt.title("ancho de spalo Vs largo de spalo")
plt.grid(True)
plt.show()
```
