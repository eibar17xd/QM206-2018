```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fraccion(ph,pka):
    
    H=10**(-ph)
    ka=[1]
    for pk in pka:
        x=10**(-pk)
        ka.append(x)
    n=len(ka)-1
    mult=1
    terminos=[]
    for a in range(n+1):
        n1=(n-a)
        cH=H**n1
        mult=mult*ka[a]
        d=cH*mult              #d=denominador
        terminos.append(d)
    sumaterminos=sum(terminos)
    alpha=terminos/sumaterminos
    return alpha
```

```python
pka=np.array([4.0,6.7,10.3])
ph=0
x=[]
y0=[]
y1=[]
y2=[]
y3=[]
while ph<=14:
    x.append(ph)
    alpha=fraccion(ph,pka)
    y0.append(alpha[0])
    y1.append(alpha[1])
    y2.append(alpha[2])
    y3.append(alpha[3])

    ph+=0.1
plt.plot(x,y0)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.ylabel("fracciones de las especies")
plt.xlabel("ph")
plt.title("fracciones ácidas")
plt.show()
```
